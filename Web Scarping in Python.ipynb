```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
html_target = "a"
tag = "src"
f_ext = ".jpg"
dir_name = "CS-Berekeley images"
url = "https://www.cs.berkeley.edu"
```

```python
r = requests.get(url)

status = r.status_code
encoding = r.encoding
html_doc = r.text

soup = BeautifulSoup(html_doc, 'html.parser')
anchor = soup(html_target)
```

```python
type(anchor)
```

```python
def make_dir(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
def download_url(url, endpoint):
    url_addr = "{url}/{endpoint}".format(url=url, endpoint=endpoint)
    file_path = "{directory}/{endpoint}".format(directory=dir_name, endpoint=endpoint)
    
    r = requests.get(url_addr)
    content_file = r.content
    
    with open(file_path, 'wb') as f:
        print("Downloading From: {url}\nWriting to: {file_path}".format(url=url_addr, file_path=file_path))
        f.write(content_file)
```

```python
print("Status: {status}\nEndoding: {encoding}".format(status=status, encoding=encoding))
print("Begin downloading")

make_dir(dir_name)
for a in anchor:
    endpoint = a[tag]
    if endpoint[-4:] == f_ext:
        download_url(url, endpoint)
        print("Finished Download -- {tag}".format(tag=endpoint))
print("Finished Downloading")
```

<h1 align = "center">CBIT Web Scraping</h1>

### 1.Importing the Libraries

```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
cbit_url = "http://www.cbit.ac.in"
```

### 2.Creating a function which takes the url and returns the HTML page source

```python
def html_page_source(url):
    uClient = uReq(url)
    html_page = uClient.read()
    uClient.close()
    return html_page
```

```python
cbit_page_html = html_page_source(cbit_url)
```

### 3.Function for creating the soup of html page source and extracting menu and their links in a dictionary

```python
def menu_and_link_dictionary(html_page):
    page_soup = soup(html_page, "html.parser")
    menu_containers = page_soup.findAll("ul", {"class":"menu"})
    menu = menu_containers[0]
    menu1 = menu.findAll("li")
    
    menu_links_dict = {}
    for items in menu1:
        link = items.a['href']
        menu_item = items.a.text
        if link[0] == '/':
            menu_links_dict[menu_item] = cbit_url + link
        elif link[0] == 'h':
            menu_links_dict[menu_item] = link
    
    return menu_links_dict
```

```python
links_dict = menu_and_link_dictionary(cbit_page_html)
```

### 4.Fuction for getting the departments and their Links

```python
def dept_and_links_dictionary(dept_url):
    cbit_departments_page_html = html_page_source(dept_url)
    cbit_departments_page_html_soup = soup(cbit_departments_page_html, "html.parser")
    departments_containers = cbit_departments_page_html_soup.findAll("div",{"class":"view-item-Departments"})
    
    dept_links_dict = {}
    for container in departments_containers:
        depart_name = container.div.text
        link = container.div.a['href']
    
        dept_links_dict[depart_name] = cbit_url + link
    return dept_links_dict
```

```python
dept_links_dict = dept_and_links_dictionary(cbit_departments_url)
```
